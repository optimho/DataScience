```python
import pandas as pd
import seaborn as sns
```

```python
frame = pd.read_csv("data/ex5.csv")
frame
```

```python
sns.barplot(data = frame)
```

```python
sns.barplot(data=frame[['sale1','sale2','sale3']])
```

```python
sns.barplot(data=frame[['sale1','sale2','sale3']],ci=False)
```

```python
sns.barplot(x= frame.year, y= frame.sale1, ci=False)
```

```python
sns.barplot(x = 'year', y = 'sale2', data = frame)
```

```python
sns.barplot(x = frame['sale3'], y=frame['year'], data = frame, orient='h', ci=False)
```

```python
sns.barplot(x = frame.year, y= frame.sale1, data = frame, hue = frame.quarters)
```
