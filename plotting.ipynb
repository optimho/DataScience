```python
data = np.random.randn(200).cumsum()
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(data)
ticks = ax.set_xticks(range(0,250,50))
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(data)
ticks = ax.set_xticks(range(0,250,50))
labels = ax.set_xticklabels(['one','two','three','four','five'])


#Make sure the set_xticks has the same amount of elements as the labels - 5 elements in this case
```

```python

fig, ax = plt.subplots(1,1)
ax.plot(data)
ticks = ax.set_xticks(range(0,250,50))
labels = ax.set_xticklabels(['year one','year two','year three','year four','year five'], rotation=60)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(data)
ticks = ax.set_xticks(range(0,250,50))
labels = ax.set_xticklabels(['year one','year two','year three','year four','year five'], rotation=60)
ax.set_xlabel('Cumalative sum')
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(data)
ticks = ax.set_xticks(range(0,250,50))
labels = ax.set_xticklabels(['year one','year two','year three','year four','year five'], rotation=60)
ax.set_xlabel('Cumalative sum')
ax.set_title('Cumulative Sum for Random Number ')
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(data)
ticks = ax.set_xticks(range(0,250,50))
labels = ax.set_xticklabels(['year one','year two','year three','year four','year five'], rotation=60)
ax.set_xlabel('Cumalative sum')
ax.set_ylabel('Value')
ax.set_title('Cumulative Sum for Random Number ')
```

```python
data1 = np.random.randn(200).cumsum()
data2 = np.random.randn(200).cumsum()
data3 = np.random.randn(200).cumsum()
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(data1)
ax.plot(data2)
ax.plot(data3)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(data1, label='data1')
ax.plot(data2, label='data2')
ax.plot(data3, label='data3')
ax.legend(loc='best')
```

```python
x = np.arange(0,10,0.1)
y=np.sin(x)
```

```python
fig, ax=plt.subplots(1,1)
ax.plot(x,y)
ax.text(4,0.7,'y=sin(x)', fontsize=15)
```

```python
# if you want to use the coordinates relative to the chart, 0.5, 0.5 is the middle
fig, ax=plt.subplots(1,1)
ax.plot(x,y)
ax.text(0.5,0.7,'y=sin(x)', fontsize=15, transform=ax.transAxes)
```

```python
fig, ax=plt.subplots(1,1)
ax.plot(x,y)
ax.text(0.5,0.7,'y=sin(x)', fontsize=15, transform=ax.transAxes)
ax.arrow(0.5,0.5, 0.6,0.5, width=0.05)
ax.arrow(6.3,0.75, 0.6,0, width=0.05)
```

```python
x = np.arange(6)
y=np.array([23,34,65,78,51,55])
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(x,y, 'ko--')
ax.annotate('max', xy=(x[3], y[3]), xytext=(3,50), arrowprops=dict(facecolor='green', shrink = 0.1))
ax.annotate('minimum', xy=(x[0], y[0]), xytext=(1,25), arrowprops=dict(facecolor='yellow', shrink=0))
```

```python
fig, ax = plt.subplots(1,1)
circle = plt.Circle((0.5,0.5),0.2, fc='y')
rect = plt.Rectangle((0.2,0.2),0.4,0.4, color='red', alpha=0.3)
ax.add_patch(circle)
ax.add_patch(rect)
```

```python
fig, ax = plt.subplots(1,1)
circle = plt.Circle((0.5,0.5),0.2, fc='y')
poly = plt.Polygon([[0.25,0.25],[0.55,0.5],[0.4,0.7]], color='red', alpha=0.5)
ax.add_patch(poly)
```
